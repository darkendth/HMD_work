```python
class PreProcessData:
    def __init__(self):
        self.malware_data = {}
        self.benign_data = {}
        self.features = ['0x04000', '0x02700', '0x02902', '0x02B00', '0x02904', '0x03000']
    
    def load_data(self):
        benign_files = ['amd_desktop_metadata/benign_post_pca_data.txt',\
                'amd_python_benign_metadata/python_benign_post_pca_data.txt'
                ]
        malware_file = 'amd_malware_metadata/pca_data_dict.txt'

        with open(malware_file, 'r') as data_file:
            mlw_data = json.load(data_file)
        
        # remove malware whose feature set are not matched and 
        # normalized the malware on over 32 run.

        self.malware_data = mlw_data

        sample_data = {}
        for f in self.features:
            sample_data[f]=[np.zeros(32)]

        el = []
        for k in mlw_data:
            mlw = mlw_data[k]
            sd = sample_data
            for s in mlw:
                sample = mlw[s]
                for f in sample:
                    if f not in sd:
                        if k not in el:
                            el.append(k)
                    else:
                        sd[f] = sd[f]+np.array(sample[f])

            self.malware_data[k]=sd
        
        for m in el:
            del self.malware_data[m]
        
        for m in self.malware_data:
            for s in self.malware_data[m]:
                self.malware_data[m][s] = pd.DataFrame(self.malware_data[m][s])


    def normalized_data(self):
        for m in self.malware_data:
            mlw = self.malware_data[m]
            for f in mlw:
                mlw[f] = mlw[f].astype(float)
                mlw[f] = np.divide(mlw[f],32)
                
        
```

```python
benign_files = ['amd_desktop_metadata/benign_post_pca_data.txt',\
        'amd_python_benign_metadata/python_benign_post_pca_data.txt'
        ]
malware_files = ['amd_malware_metadata/malware_post_pca_data.txt']

benign_dict={}
benign_data=[]

for files in benign_files:
    with open(files, 'r') as data_file:
        json_data = json.load(data_file)

    for bm in json_data.keys():
        benign_dict[bm]        = False
        for each_sample in json_data[bm]:
            if benign_data == []:
                benign_data = np.array(each_sample) 
            else:
                benign_data = np.vstack((benign_data, each_sample))
    
    data_file.close() 
```

```python
malware_files = ['amd_malware_metadata/malware_post_pca_data.txt']

malware_dict = {}
malware_data = []

for files in malware_files:
    with open(files, 'r') as data_file:
        json_data = json.load(data_file)

    for bm in json_data.keys():
        malware_dict[bm]       = False
        for each_sample in json_data[bm]:
            if malware_data == []:
                malware_data = np.array(each_sample) 
            else:
                try:
                    malware_data = np.vstack((malware_data, each_sample))
                except:
                    continue

    data_file.close()
```

```python
benign_label   = np.zeros(len(benign_data))
malware_label  = np.ones(len(malware_data))

sample_size    = min(len(benign_label), len(malware_label))

data           = np.vstack((\
                        random.sample(list(benign_data), sample_size),\
                        random.sample(list(malware_data), sample_size)\
                        ))
label          = np.concatenate((\
                        random.sample(list(benign_label), sample_size),\
                        random.sample(list(malware_label), sample_size)\
                        ))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_fscore_support
```

```python
X_train, X_test, y_train, y_test = train_test_split(\
                            data, label, test_size = 0.1, \
                            random_state = random.randint(0, 2**32 -1 ))
```
